# Question 9
Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

roman_file = pd.read_csv("roman-emperor-reigns.csv")

death_cause = roman_file.groupby('Cause_of_Death')['Emperor'].count()

death_cause.plot.pie(y='Cause_of_Death', figsize=(8, 8), autopct='%1.1f%%', startangle=0, textprops={'fontsize': 11})

plt.title('Roman Emperors that were assassinated')
plt.ylabel('')
```
