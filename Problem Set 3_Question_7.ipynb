# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt
file = pd.read_csv('us-marriages-divorces-1867-2014.csv')

marry = file[(file.Year == 2000) | (file.Year == 1900) | (file.Year == 1950)]
marry.plot(kind= 'bar', x='Year', y=["Marriages_per_1000", "Divorces_per_1000"])
plt.title("Marriages and Divorces between 1900, 1950 and 2000")
plt.ylabel("Marriages & Divorces")
plt.show()
```
