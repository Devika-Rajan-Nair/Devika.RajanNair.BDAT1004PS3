# Question 6
Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. 
Label both lines and show the legend.
Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
import pandas as pd

file = pd.read_csv("us-marriages-divorces-1867-2014.csv")

fig, ax = plt.subplots(figsize=(10,5))

plt.plot(file.Year, file['Marriages'], label='Marriages per capita')
plt.plot(file.Year, file['Divorces'], label='Divorces per capita')

plt.title("Marriages and Divorces in U.S between 1867-2014")
plt.xlabel(" Year")
plt.ylabel(" Number of Marriages /Divorces");

plt.legend();
```
